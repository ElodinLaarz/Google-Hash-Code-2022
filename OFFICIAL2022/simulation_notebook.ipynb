```python
from gz_parser.parser import Parser
from simulation import Simulator

# 
prefixes = ['a', 'b', 'c', 'd', 'e', 'f']

for prefix in prefixes:

    p = Parser('./inputs/' + prefix + '.in')

    current_simulation = Simulator(p, prefix)

    print(f'{prefix} has {len(current_simulation.contributors)} members.')

    counter = 0
    hard_stop = len(current_simulation.projects)*2+3
    while (not current_simulation.done and counter < hard_stop):
        current_simulation.time_step()
        counter += 1

    if counter == hard_stop:
        print('We had to hard_stop.')

    current_simulation.create_output()
```
